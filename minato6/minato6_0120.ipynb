```python
ap = minato6ap[['datetime_timestamp','CSI_DATA','len']]
mp = minato6mp[['datetime_timestamp','CSI_DATA','len']]
ap_timestamp = ap['datetime_timestamp'].to_numpy()
mp_timestamp = mp['datetime_timestamp'].to_numpy()
```

```python
ap.dtypes
```

```python
mp.dtypes
```

```python
idx = np.searchsorted(ap_timestamp, mp_timestamp) - 1
mask = idx >= 0

df = pd.DataFrame({"ap_timestamp":ap_timestamp[idx][mask], "mp_timestamp":mp_timestamp[mask]})
```

```python
df.head()
```

```python
df.tail()
```

timestampが14:02 - 15:10 のもの以外は消す
15:00 - 15:10 のものは分ける。

```python
df
```

```python
ap.columns = ['ap_timestamp', 'CSI_ap','len_ap']
mp.columns = ['mp_timestamp', 'CSI_mp','len_mp']
join_data_1 = pd.merge(df, ap, on='ap_timestamp', how='left')
join_data_2 = pd.merge(join_data_1, mp, on='mp_timestamp', how='left')
join_data = join_data_2[['ap_timestamp', 'mp_timestamp', 'CSI_ap','CSI_mp','len_ap','len_mp',]]
join_data.isnull().any(axis=0)
join_data.head(20)
```

```python
#print(join_data['CSI_ap'].dtypes,type(join_data['CSI_ap'].iloc[0]))
print(join_data['CSI_ap'].iloc[0][1:-2])
```

```python
join_data['CSI_ap_int'] = join_data['CSI_ap'][1:-2]
join_data['CSI_ap_int'].iloc[100]
```

```python
join_data['CSI_ap_arr'] = join_data['CSI_ap'].apply(lambda x: str(x)[1:-2].split(" "))
```

```python
join_data['CSI_ap_arr'].head(20)
```

```python
#could not convert string to float: '-' エラーがあったので、"-"どこにあるか探してみる,他の異常値の場合も、ここでフィルタ処理

for i,each in enumerate(join_data['CSI_ap_arr']):
  if "-" in each:
    print(i)
    print(join_data['CSI_ap_arr'].iloc[i])
    tar_arr=join_data['CSI_ap_arr'].iloc[i]
    for j,ele in enumerate(tar_arr):
      if ele=="-":
        print("found!! - at", j,"-th element")
        tar_arr[j] = "0"#"-"を"0"に変更
        print(tar_arr)
        join_data['CSI_ap_arr'].iloc[i]=tar_arr[j]
```

```python
join_data['CSI_ap_arr_int']=join_data['CSI_ap_arr'].apply(lambda x: np.array(x).astype(np.float))
```

```python
join_data['CSI_ap_arr_int'].head(20)
```

```python
join_data['CSI_ap_arr_mean']=join_data['CSI_ap_arr_int'].apply(lambda x:np.mean(x))
join_data['CSI_ap_arr_max']=join_data['CSI_ap_arr_int'].apply(lambda x:np.max(x))
join_data['CSI_ap_arr_min']=join_data['CSI_ap_arr_int'].apply(lambda x:np.min(x))
```
